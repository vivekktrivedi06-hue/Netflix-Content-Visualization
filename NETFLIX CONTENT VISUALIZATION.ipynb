```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df.fillna("Unknown", inplace=True)
```

```python
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries")
plt.show()
```

```python
sns.countplot(x='rating', data=df, order=df['rating'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Ratings Distribution")
plt.show()
```

df['release_year'].value_counts().sort_index().plot()
plt.title("Content Release Over Years")
plt.show()

```python
df['director'].value_counts().head(10).plot(kind='barh')
plt.title("Top Directors")
plt.show()
```

Netflix has more Movies than TV Shows
Most content comes from USA
Content increased rapidly after 2015
Majority ratings are TV-MA / TV-14

```python
sns.countplot(x='type', hue='rating', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
df[df['director'] != "Unknown"]['director'].value_counts().head(10).plot(kind='barh')
plt.title("Top Directors (Filtered)")
plt.show()
```
